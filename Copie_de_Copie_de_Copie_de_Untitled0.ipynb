**Segmentation des clients**

1. La répartition des ventes entre les différents segments de clients
2. Les segments de clients des préférences spécifiques en termes de produits ou de régions



```python
import pandas as pd
import matplotlib.pyplot as plt

# Charge les données dans un DataFrame (remplace "ton_fichier.csv" par le chemin de ton fichier CSV)
df = pd.read_csv("Salesstore.csv")

# Groupe par segment de client et calcule le total des ventes
segment_sales = df.groupby('Customer_Segment')['Sales'].sum().sort_values(ascending=False)

# Affiche le graphique en barres
plt.figure(figsize=(10, 6))
segment_sales.plot(kind='bar', color='salmon')
plt.title('Répartition des Ventes entre les Segments de Clients')
plt.xlabel('Segment de Client')
plt.ylabel('Ventes')
plt.show()
```

**Performance des produits**
1. les produits les plus vendus
2. les corrélations entre la catégorie ou la sous-catégorie d'un produit et ses performances



```python
import pandas as pd
import matplotlib.pyplot as plt

# Assurez-vous de charger vos données dans un DataFrame, remplacez 'votre_fichier.csv' par le chemin de votre fichier
donnees_ventes = pd.read_csv('Salesstore.csv')

# Calculer le chiffre d'affaires total par catégorie de produit
ventes_par_categorie = donnees_ventes.groupby('Product_Category')['Sales'].sum().reset_index()

# Trier les résultats par chiffre d'affaires (du plus au moins élevé)
categories_plus_populaires = ventes_par_categorie.sort_values(by='Sales', ascending=False)

# Afficher les résultats sous forme de graphique à barres
plt.figure(figsize=(12, 6))
plt.bar(categories_plus_populaires['Product_Category'], categories_plus_populaires['Sales'], color='green')
plt.title('Catégories de produits les plus populaires en termes de ventes')
plt.xlabel('Catégorie de produit')
plt.ylabel('Chiffre d\'affaires total')
plt.xticks(rotation=45, ha='right')  # Rotation des étiquettes pour une meilleure lisibilité
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assurez-vous de charger vos données dans un DataFrame, remplacez 'votre_fichier.csv' par le chemin de votre fichier
donnees_ventes = pd.read_csv('Salesstore.csv')

# Calculer la somme des quantités vendues par produit
quantites_par_produit = donnees_ventes.groupby('Product_Name')['Order_Quantity'].sum().reset_index()

# Calculer le chiffre d'affaires total par produit
chiffre_affaires_par_produit = donnees_ventes.groupby('Product_Name')['Sales'].sum().reset_index()

# Fusionner les résultats
resultats = pd.merge(quantites_par_produit, chiffre_affaires_par_produit, on='Product_Name', how='inner')

# Renommer les colonnes
resultats.columns = ['Product_Name', 'Total_Quantity_Sold', 'Total_Sales']

# Trier les résultats par quantité vendue (du plus au moins vendu)
produits_plus_vendus_quantite = resultats.sort_values(by='Total_Quantity_Sold', ascending=False).head(10)

# Trier les résultats par chiffre d'affaires (du plus au moins élevé)
produits_plus_vendus_chiffre_affaires = resultats.sort_values(by='Total_Sales', ascending=False).head(10)

# Afficher les résultats sous forme de graphique à secteurs
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 14))

# Graphique à secteurs pour les produits les plus vendus en termes de quantité
axes[0].pie(produits_plus_vendus_quantite['Total_Quantity_Sold'], labels=produits_plus_vendus_quantite['Product_Name'], autopct='%1.1f%%', startangle=90, colors=plt.cm.viridis.colors)
axes[0].set_title('Top 10 des produits les plus vendus en termes de quantité')

# Graphique à secteurs pour les produits les plus vendus en termes de chiffre d'affaires
axes[1].pie(produits_plus_vendus_chiffre_affaires['Total_Sales'], labels=produits_plus_vendus_chiffre_affaires['Product_Name'], autopct='%1.1f%%', startangle=90, colors=plt.cm.viridis.colors)
axes[1].set_title('Top 10 des produits les plus vendus en termes de chiffre d\'affaires')

# Ajuster la disposition
plt.tight_layout()

# Afficher les graphiques à secteurs
plt.show()
```
